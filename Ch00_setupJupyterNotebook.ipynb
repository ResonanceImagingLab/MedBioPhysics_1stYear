## üìä A Preview: Why We Use Code

Here is an example of something we *cannot* easily do with a calculator. Note, the first line containing `%pip` is used to install external python packages. This one in particular is used for plotting.


```python
%pip install matplotlib 
import matplotlib.pyplot as plt

time = [0, 1, 2, 3, 4]
temperature = [37, 37.2, 37.5, 38, 38.6]

plt.plot(time, temperature)
plt.xlabel("Time (hours)")
plt.ylabel("Body Temperature (¬∞C)")
plt.title("Example: Temperature Over Time")
plt.show()
```
