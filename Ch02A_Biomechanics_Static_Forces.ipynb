## ü¶µ Part 1 ‚Äî Standing on One Leg: Force Balance

When you stand still, your body is **not accelerating**.
This means forces must balance.

We start with a **simplified model**:
- One leg supports the body
- Gravity pulls downward
- The knee joint provides an upward force


```python
# Body parameters
body_mass = 75      # kg
g = 9.81            # m/s^2

body_weight = body_mass * g
body_weight
```

This is the **force due to gravity** acting on the body (in Newtons).

If one leg supports the body, the knee joint must supply an equal upward force.


```python
knee_force = body_weight
knee_force
```

## ‚öñÔ∏è Part 2 ‚Äî Balance and Stability

Balance depends on where the **center of mass** is relative to the feet.

If the center of mass moves too far forward or backward, balance is lost.

We model the foot as a support region.


```python
# Simple balance model (1D)
foot_length = 0.25   # meters
center_of_mass = 0.15  # meters from heel

stable = 0 <= center_of_mass <= foot_length
stable
```

`True` means stable balance.

### Try this

You should find that once the center of mass is outside the foot length, that the model is no longer stable. This should make sense as you can do a reasonable amount of leaning forward before stumbling.

## ü¶æ Part 3 ‚Äî Levers in the Human Body

Bones act as **levers**, and joints act as **pivot points**.

There are three classes of levers:
1. First class (see-saw)
2. Second class (wheelbarrow)
3. Third class (most human joints)

The elbow is a **third-class lever**.


```python
# Lever geometry for elbow
load_force = 50       # N (weight in hand = mass*gravity)
load_distance = 0.35  # m from elbow
muscle_distance = 0.04  # m from elbow - this is muscle insertion

muscle_force = load_force * load_distance / muscle_distance
muscle_force
```

Even a small weight requires a **large muscle force**.

This close connection of a muscle to the joint shows how the class 3 lever results in increased forces required to lift something.

## ü¶µ Part 4 ‚Äî Knee Joint Forces (External Load Model)

We begin with a **simplified external-force model** of the knee during quiet standing.

In this model, we consider only:
- Body weight acting downward
- Ground reaction force acting upward
- Static equilibrium (no acceleration)

At this stage, we **do not explicitly model muscle forces**.
This allows us to focus on basic force balance before introducing internal loads.

### Force balance during quiet standing

Because the person is standing still, the acceleration is zero.
Therefore, the **net force must be zero**:

$$
\sum F = 0
$$

For the vertical (up‚Äìdown) direction, this becomes:

$$
F_{\text{ground}} - W = 0
$$

or equivalently,

$$
F_{\text{ground}} = W
$$

where:
$$
W = \text{person's weight}
$$

$$
F_{\text{ground}} = \text{ground reaction force}
$$


### What does the ‚Äúreaction force‚Äù mean here?

In this simplified model, the **ground reaction force** represents the force exerted by the ground on the feet.

Because forces must transmit through the body, this same force is passed upward through the leg and across the knee joint.

In other words:
- The knee must transmit a force equal to body weight
- This force is sometimes called the **joint reaction force** in a simplified sense
- At this level, it represents the force required to support the body, not the detailed internal stresses within the joint

### Adding muscle force: a conceptual extension

In real joints, muscles must generate force to maintain posture and stability.
These muscle forces act **in addition to external loads** and contribute to the
total force transmitted through a joint.



```python
# Body mass (kg)
body_mass = 75  

# Gravitational acceleration (m/s^2)
g = 9.81  

# Body weight (N)
body_weight = body_mass * g


# Example quadriceps muscle force (N)
# This value is illustrative and represents internal muscle loading to maintain posture
quadriceps_force = 1500  

# --- Joint force transmission ---

# The total force transmitted through the knee joint includes:
# 1) External load (body weight)
# 2) Internal muscle force (quadriceps)

total_knee_force = body_weight + quadriceps_force

body_weight, quadriceps_force, total_knee_force
```

### Interpretation
Joint forces can be **several times body weight**, even without motion. We will want to keep this in mind in the next chapter.

## ü¶¥ Part 5 ‚Äî Bone Stress: Compression vs Tension

Stress describes how force is distributed over an area.

Stress = Force / Area


```python
# Bone cross-sectional area
bone_area = 3e-4  # m^2

compressive_stress = total_knee_force / bone_area
compressive_stress
```

This is **compressive stress** in the bone.

Bones are generally stronger in compression than in tension.
