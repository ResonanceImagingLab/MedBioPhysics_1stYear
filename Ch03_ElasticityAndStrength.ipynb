## ü¶¥ Part 1 ‚Äî Stress vs. Strain

When a force is applied to bone, two related quantities matter:

- **Stress**: how much force is applied per area  
- **Strain**: how much the bone deforms relative to its original size

Stress causes strain.


```python
# Force applied to bone
force = 3000   # N
area = 3       # cm^2

# cm^2 makes more intuitive sense when thinking about bone size, but we need to convert to m^2 for S.I. unit purposes
area = area * 0.0001

stress = force / area
stress
```

Stress is measured in **Pascals (Pa)**, where 1 Pa = 1 N/m $^2$ 

Now consider bone deformation.


```python
# Bone length change
original_length = 0.4   # m
delta_length = 0.001    # m

strain = delta_length / original_length
strain
```

Strain has **no units** ‚Äî it is a ratio.

### Key idea
Stress describes *loading*.  
Strain describes *response*.


---
## üîó Part 2 ‚Äî Elasticity and Hooke‚Äôs Law

In the elastic regime, stress and strain are proportional.

This relationship is known as **Hooke‚Äôs Law**:


```python
# Young's modulus for bone (approximate)
youngs_modulus = 17e9  # Pa - or 1.7GigaPascal (GPa)

predicted_stress = youngs_modulus * strain
predicted_stress
```

Young‚Äôs modulus describes the **stiffness** of a material. 

Higher modulus ‚Üí stiffer material. If we think of a spring, the stiffer the spring, the more it resists deformation.


---
## üìà Part 3 ‚Äî Stress‚ÄìStrain Curve

We now visualize how stress changes with strain.


```python
%matplotlib inline
import matplotlib.pyplot as plt

strain_values = [0, 0.001, 0.002, 0.003, 0.004]
stress_values = [youngs_modulus * s for s in strain_values] # this applied the formula to all strain values

strain_valuesPercent = [100 * s for s in strain_values]

plt.plot(strain_valuesPercent, stress_values)
plt.xlabel("Strain (%)")
plt.ylabel("Stress (Pa)")
plt.title("Elastic Stress‚ÄìStrain Relationship for Bone")
plt.show()
```

The straight line indicates **elastic behavior**. Note that the strain ratio has been converted to a percentage, as this describes the percent change in length. This is achieved by multiplying the the strain ratio by 100 and provides a more intuitive value to compare.

Under elastic conditions, if the load is removed, the bone returns to its original shape.

## üß± Part 4 ‚Äî Compression vs. Tension

Bones behave differently under:

- **Compression** (pushing)
- **Tension** (pulling)

They are generally stronger in compression. This results in a different Youngs Modulous for each action. These values are determined empirically (experimentally).


```python
compressive_strength = 170e6  # Pa
tensile_strength = 130e6       # Pa

compressive_strength, tensile_strength
```

### Interpretation
Bone can tolerate **higher compressive stress** than tensile stress.

## üîÄ Part 5 ‚Äî Shear Stress

Shear stress occurs when forces act **parallel** to a surface (or perpendicular to the primary axis). Think of a deck of cards on a table, and you push the deck sideways causings the cards to slide parallel to each other.

```python
# Shear force example
shear_force = 1200   # N
shear_area = 3       # cm^2 - cross section of object parallel to applied shear force

# cm^2 makes more intuitive sense when thinking about bone size, but we need to convert to m^2 for S.I. unit purposes
shear_area = shear_area * 0.0001

shear_stress = shear_force / shear_area
shear_stress
```

The equation replicates the non-shear stress equation, being a force divided by an area. 
It can also be calculated by using the Shear Modulous (replaces Young's modulus) and the angle of deflection œÜ in radians.

```python
deflection_degrees = 5
shear_modulous = 80e9

deflection_radians = deflection_degrees *3.1416 /180

shear_stress = shear_modulous *deflection_radians

shear_stress
```

Under elastic conditions, you can observe that shear stress increases linearly with deflection angle.

This continues until the conditions for elasticity are no longer met, which can occur at high shear stresses. Shear stress is often involved in **spiral and oblique fractures**.

Bones are well designed to support load along their primary (longitudinal) axis, where compressive stresses dominate and the material properties of bone are well matched to the applied forces. When loads are applied at an angle, however, the internal stress state changes dramatically.

At first glance, this may seem counterintuitive, since the Young‚Äôs modulus of bone is often lower than its shear modulus, suggesting that bone should fail more readily in compression than in shear. Because bones act as long lever arms, even modest forces applied at a distance from a joint generate large torques (œÑ = r √ó F). This torque increases bending moments along the shaft, leading to elevated shear stresses and tensile stresses that bone is comparatively poor at resisting. As a result, fractures often occur not because the applied force is large, but because the effective moment arm is long, converting a moderate force into a large internal stress.

In this sense, shear stress and torque are central to understanding bone fractures: it is not just how hard a bone is loaded, but where and how the force is applied. This explains why falls, twisting injuries, and off-axis impacts are far more likely to cause fractures than purely axial loading, even when the total applied force is similar.
