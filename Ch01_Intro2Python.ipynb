## üß± Part 1 ‚Äî Variables: Storing Information

A **variable** is a labeled box that stores a value.

Run the cell below.


````python
heart_rate = 72
heart_rate
````

You can change the value and run the cell again.


````python
heart_rate = 65
heart_rate
````

Variables hold only the last value they were set to, and can be used in calculations.


````python
beats_per_minute = heart_rate
beats_per_hour = beats_per_minute * 60
beats_per_hour
````

## üì¶ Part 3 ‚Äî Lists: Working with Multiple Values

In science, we often work with **many values**, not just one.

A **list** stores multiple values together. In programming, the specific bracket type you use is important. For lists of numbers and indexing in python, use square brackets [...].


````python
time = [0, 1, 2, 3, 4]
time
````

Lists can store measurements.


````python
temperature = [37.0, 37.2, 37.5, 38.0, 38.6]
temperature
````

Each position in a list has an index (starting at 0, not 1).


````python
time[0], time[1], temperature[0], temperature[1]
````

## üìä Part 4 ‚Äî Plotting Data (A Pattern You Will Reuse)

We will use the same plotting pattern throughout the course. 

If you get errors here related to matplotlib - please see Ch00_setupJupyterNotebook.ipynb and run the cell under the heading 'A Preview: Why We Use Code'


````python
%matplotlib inline
import matplotlib.pyplot as plt
````

````python
plt.plot(time, temperature)
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (¬∞C)")
plt.title("Body Temperature Over Time")
plt.show()
````
